```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_bvp
```

```python
def prob1(y, t):
    dSdt = -1/2 * y[1]*y[0]
    dIdt = 1/2 * y[1]*y[0] - 1/4 * y[1]
    dRdt = 1/4 * y[1]
    return dSdt, dIdt, dRdt

t = np.linspace(0, 100, 101)
y0 = [1 - 6.25e-7, 6.25e-7, 0]
sol = odeint(prob1, y0, t)
plt.plot(t, sol[:,0], label="Susceptible")
plt.plot(t, sol[:,1], label="Infected")
plt.plot(t, sol[:,2], label="Recovered")
plt.xlabel("T (days)")
plt.ylabel("Proportion of Population")
plt.legend()
plt.show()
```

```python
def prob2(y, t):
    dSdt = -1 * y[1]*y[0]
    dIdt = 1 * y[1]*y[0] - 1/3 * y[1]
    dRdt = 1/3 * y[1]
    return dSdt, dIdt, dRdt

t = np.linspace(0, 50, 101)
y0 = [1 - 1.67e-6, 1.67e-6, 0]
sol = odeint(prob2, y0, t)
plt.plot(t, sol[:,0], label="Susceptible")
plt.plot(t, sol[:,1], label="Infected")
plt.plot(t, sol[:,2], label="Recovered")
plt.xlabel("T (days)")
plt.ylabel("Proportion of Population")
plt.legend()
plt.show()

w = max(sol[:,1])
num_ppl = (1 - w) * 3e6
print("Number of people who can still work: {}".format(int(num_ppl)))
```

```python
def prob3(y, t):
    dSdt = -3/10 * y[1]*y[0]
    dIdt = 3/10 * y[1]*y[0] - 1/4 * y[1]
    dRdt = 1/4 * y[1]
    return dSdt, dIdt, dRdt

t = np.linspace(0, 500, 101)
y0 = [1 - 1.67e-6, 1.67e-6, 0]
sol = odeint(prob3, y0, t)
plt.plot(t, sol[:,0], label="Susceptible")
plt.plot(t, sol[:,1], label="Infected")
plt.plot(t, sol[:,2], label="Recovered")
plt.xlabel("T (days)")
plt.ylabel("Proportion of Population")
plt.legend()
plt.show()
```

```python
def prob4(y, t):
    dSdt = 1/50 * y[2] + .0001*(1 - y[0]) - 1/2 * y[0]*y[1]
    dIdt = 1/2 * y[0]*y[1] - (1/10 + .0001)*y[1]
    dRdt = -1/50 * y[2] + 1/10 * y[1] - .0001*y[2]
    return dSdt, dIdt, dRdt

t = np.linspace(0, 365, 366)
y0 = [1 - 1.43e-7, 1.43e-7, 0]
sol = odeint(prob4, y0, t)
plt.plot(t, sol[:,0], label="Susceptible")
plt.plot(t, sol[:,1], label="Infected")
plt.plot(t, sol[:,2], label="Recovered")
plt.xlabel("T (days)")
plt.ylabel("Proportion of Population")
plt.legend()
plt.show()
```
