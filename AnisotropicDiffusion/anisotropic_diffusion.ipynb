# Anisotropic Diffusion

```python
import numpy as np
from matplotlib import cm, pyplot as plt
from scipy.misc import imsave
from imageio import imread
from numpy.random import randint
```

```python
from matplotlib import cm, pyplot as plt
from imageio import imread
# To read in an image, convert it to grayscale, and rescale it.
picture = imread('balloon.jpg', as_gray=True) * 1./255
# To display the picture as grayscale
plt.imshow(picture, cmap=cm.gray)
plt.show()
```

## Problem 1

```python
def anisdiff_bw(U, N, lambda_, g):
    """ Run the Anisotropic Diffusion differencing scheme
    on the array U of grayscale values for an image.
    Perform N iterations, use the function g
    to limit diffusion across boundaries in the image.
    Operate on U inplace to optimize performance. """
    U1 = U.copy()
    for _ in range(N):
        U_ = np.pad(U1, 1, "constant", constant_values=0)
        U0 = U_[1:-1,1:-1]
        A = U_[:-2,1:-1] - U0
        B = U_[2:,1:-1] - U0
        C = U_[1:-1,:-2] - U0
        D = U_[1:-1,2:] - U0
        U1 += lambda_ * (g(np.abs(A))*A
                       + g(np.abs(B))*B
                       + g(np.abs(C))*C
                       + g(np.abs(D))*D)
    return U1
```

```python
lambda_ = .25
sigma = 0.1
g = lambda x: np.exp(-(x/sigma)**2)
U5 = anisdiff_bw(picture, 5, lambda_, g)
U20 = anisdiff_bw(picture, 20, lambda_, g)
U100 = anisdiff_bw(picture, 100, lambda_, g)
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,24))
ax[0, 0].axis("off")
ax[0, 0].imshow(picture, cmap=cm.gray)
ax[0, 0].set_title("Original Image")
ax[0, 1].axis("off")
ax[0, 1].imshow(U5, cmap=cm.gray)
ax[0, 1].set_title("N = 5")
ax[1, 0].axis("off")
ax[1, 0].imshow(U20, cmap=cm.gray)
ax[1, 0].set_title("N = 20")
ax[1, 1].axis("off")
ax[1, 1].imshow(U100, cmap=cm.gray)
ax[1, 1].set_title("N = 100")

plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```

```python
from matplotlib import cm, pyplot as plt
from imageio import imread
# To read in an image, convert it to grayscale, and rescale it.
picture = imread('balloons_color.jpg') * 1./255
# To display the picture as grayscale
plt.imshow(picture)
plt.show()
```

## Problem 2

```python
def anisdiff_color(U, N, lambda_, sigma):
    """ Run the Anisotropic Diffusion differencing scheme
    on the array U of grayscale values for an image.
    Perform N iterations, use the function g = e^{-x^2/sigma^2}
    to limit diffusion across boundaries in the image.
    Operate on U inplace to optimize performance. """
    U1 = U.copy()
    npad = ((1, 1), (1, 1), (0, 0))
    norm = lambda x: np.sqrt(np.sum(x**2, axis=2, keepdims=True))
    for _ in range(N):
        U_ = np.pad(U1, npad, "constant", constant_values=0)
        U0 = U_[1:-1,1:-1]
        A = U_[:-2,1:-1] - U0
        B = U_[2:,1:-1] - U0
        C = U_[1:-1,:-2] - U0
        D = U_[1:-1,2:] - U0
        U1 += lambda_ * (g(norm(A))*A
                       + g(norm(B))*B
                       + g(norm(C))*C
                       + g(norm(D))*D)
    return U1
```

```python
lambda_ = .25
sigma = 0.1
g = lambda x: np.exp(-(x/sigma)**2)
U5 = anisdiff_color(picture, 5, lambda_, g)
U20 = anisdiff_color(picture, 20, lambda_, g)
U100 = anisdiff_color(picture, 100, lambda_, g)
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,24))
ax[0, 0].axis("off")
ax[0, 0].imshow(picture)
ax[0, 0].set_title("Original")
ax[0, 1].axis("off")
ax[0, 1].imshow(U5)
ax[0, 1].set_title("N = 5")
ax[1, 0].axis("off")
ax[1, 0].imshow(U20)
ax[1, 0].set_title("N = 20")
ax[1, 1].axis("off")
ax[1, 1].imshow(U100)
ax[1, 1].set_title("N = 100")
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```

## Problem 3

```python
image = imread('balloon.jpg', as_gray=True)
x, y = image.shape
for i in range(x*y//100):
    image[randint(x),randint(y)] = 127 + randint(127)
image *= 1./255
```

```python
lambda_ = .25
sigma = 0.1
g = lambda x: np.exp(-(x/sigma)**2)
U20 = anisdiff_bw(image, 20, lambda_, g)
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,24))
ax[0].axis("off")
ax[0].imshow(image, cmap=cm.gray)
ax[0].set_title("Noised original")
ax[1].axis("off")
ax[1].imshow(U20, cmap=cm.gray)
ax[1].set_title("N=20 filter of the noised image")
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```

The anisotropic diffusion does not smooth out the noise because the diffusion is minimizes blurring around the sharp edges of the image.
