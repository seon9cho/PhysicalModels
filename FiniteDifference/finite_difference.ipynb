```python
import numpy as np
import autograd.numpy as anp 
import matplotlib.pyplot as plt
from autograd import elementwise_grad as egrad
import scipy.sparse as sp
```

# Problem 1

```python
# What the solution should look like using 2 different methods of numeric differentiation
u = lambda x: anp.sin((x+anp.pi)**2 - 1)

x = np.linspace(0, 1, 10)
plt.plot(x, u(x))
plt.plot(x, egrad(u)(x))
plt.plot(x, np.gradient(u(x), x))
plt.show()
```

```python
# Function for calculating the first derivative
def grad1(f, x):
    n = len(x)
    M = sp.diags([-1, 0, 1], [-1, 0, 1], shape=(n-2, n-2))
    h = x[1] - x[0]
    L = np.zeros(n-2)
    L[0] = -f(x[0]) / (2*h)
    L[-1] = f(x[-1]) / (2*h)
    u_prime = 1/(2*h) * M@f(x[1:-1]) + L
    return u_prime
# Function for calculating the second derivative
def grad2(f, x):
    n = len(x)
    h = x[1] - x[0]
    M = sp.diags([1, -2, 1], [-1, 0, 1], shape=(n-2, n-2))
    L = np.zeros(n-2)
    L[0] = f(x[0]) / h**2
    L[-1] = f(x[-1]) / h**2
    u_2prime = 1/(h**2) * M@f(x[1:-1]) + L
    return u_2prime
# Solve and plot
u_prime = grad1(u, x)
u_2prime = grad2(u, x)
Y_hat = (1/2) * u_2prime - u_prime
plt.plot(x[1:-1], Y_hat)
plt.show()
```

## Problem 2

```python
# Define the necessary constants/functions
eps = 1/10
f = lambda x: -1 + x*0
alpha = 1
beta = 3
n = 100

# Function that solves singularly perturbed bvp
def sp_bvp(f, eps, alpha, beta, n):
    """
    Solve for a boundary value problem of the form
    εu"(x) - u'(x) = f(x), x ∈ (0,1)
    f: callable function
    eps: scalar
    alpha: scalar left boundary condition
    beta: scalar right boundary condition
    n: number of subintervals
    """
    x = np.linspace(0, 1, n)
    h = x[1] - x[0]
    
    M = sp.diags([eps + h/2, -2*eps, eps - h/2], [-1, 0, 1], shape=(n-2, n-2)).toarray()
    M_inv = np.linalg.inv(M)
    
    b = f(x[1:-1])
    b[0] -= alpha*(eps + h/2) / h**2
    b[-1] -= beta*(eps - h/2) / h**2
    
    U = h**2 * M_inv@b
    return np.hstack([alpha, U, beta])

# Solve and plot
x = np.linspace(0, 1, n)
U = sp_bvp(f, eps, alpha, beta, n)
plt.plot(x, U)
plt.show()
```

# Problem 3

```python
num_approx = 10 # Number of Approximations
N = 5*np.array([2**j for j in range(num_approx)])
h, max_error = (1.-0)/N[:-1], np.ones(num_approx-1)
# Best numerical solution, used to approximate the true solution.
# bvp returns the grid, and the grid function, approximating the solution
# with N subintervals of equal length.
num_sol_best = sp_bvp(f, eps, alpha, beta, N[-1])
for j in range(len(N)-1):
    num_sol = sp_bvp(f, eps, alpha, beta, N[j])
    max_error[j] = np.max(np.abs(num_sol-num_sol_best[::2**(num_approx-j-1)]))
plt.loglog(h,max_error,'.-r',label="$E(h)$")
plt.loglog(h,h**(2.),'-k',label="$h^{\, 2}$")
plt.xlabel("$h$")
plt.legend(loc='best')
plt.show()
print("The order of the finite difference approximation is about ",
(np.log(max_error[0])-np.log(max_error[-1]))/(np.log(h[0])-np.log(h[-1])),
".")
```

# Problem 4

```python
# Function to solve general linear 2nd order bvp
def linear_bvp(f, coef, alpha, beta, interval, n):
    """
    Solve for a boundary value problem of the form
    a1(x)y" + a2(x)y' + a3(x)y = f(x), x ∈ (a,b)
    f: callable function
    coef: list of callable functions, corresponding to [a1, a2, a3]
    alpha: scalar left boundary condition
    beta: scalar right boundary condition
    interval: (a,b)
    n = number of subintervals
    """
    x = np.linspace(interval[0], interval[1], n)
    h = x[1] - x[0]
    a1 = coef[0](x)
    a2 = coef[1](x)
    a3 = coef[2](x)
    
    A = (a1/(h**2)) + (a2/(2*h))
    B = (-2*a1/(h**2)) + a3
    C = (a1/(h**2)) - (a2/(2*h))
    
    M = np.diag(C[2:-1], k=-1) + np.diag(B[1:-1]) + np.diag(A[1:-2], k=1)
    M_inv = np.linalg.inv(M)
    
    b = f(x[1:-1])
    b[0] -= C[1]*alpha
    b[-1] -= A[-1]*beta
    
    U = M_inv@b
    return np.hstack([alpha, U, beta])
```

```python
# Defined constants/functions
f = lambda x: np.cos(x)
a1 = lambda x: 0.1 + x*0
a2 = lambda x: x*0
a3 = lambda x: -4*(np.pi - x**2)
interval = (0, np.pi/2)
alpha = 0
beta = 1
n = 100
# Solve and plot
U = linear_bvp(f, [a1,a2,a3], alpha, beta, interval, n)
x = np.linspace(interval[0], interval[1], n)
plt.plot(x, U)
plt.show()
```
