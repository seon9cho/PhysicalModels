```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

# Inverse Problems

## Problem 1
Solve (19.2) for $a(x)$ using the following conditions:

$c_0 = 3/8,$ $c_1 = 5/4$, $u(x) = x^2 + x/2 + 5/16$, $x_j=.1j$ for $j=0,1,\dots,10$, and 
\begin{align*}
	f &= \begin{cases}
		-6x^2 + 3x - 1 & x \leq 1/2,\\
		-1 & 1/2 < x \leq 1,
	\end{cases}
\end{align*}
Produce the plot shown in Figure 19.2.

Hint: use the `minimize` function in `scipy.optimize` and some initial guess to find the $a_j$.

```python
c0 = 3/8
c1 = 5/4
x = np.linspace(0, 1, 11)
du_func = lambda x: 2*x + 1/2
du = du_func(x)
def f_func(x):
    if x <= 1/2:
        return -6*x**2 + 3*x - 1
    else:
        return -1
If = np.array([quad(f_func, 0, xj)[0] for xj in x])

def F(a):
    return np.sum(((c0 - If)/a - du)**2)

guess = np.ones(len(x))*1/2
a = minimize(F, guess).x
plt.plot(x, a)
plt.show()
```

## Problem 2
Find the density function $a(x)$ satisfying 
	\begin{align}
	\begin{cases}
		-(au')' = -1, & x \in (0,1),\\
		a(0)u'(0) = 1, & a(1)u'(1) = 2.
	\end{cases} \label{inverse_problems:ill_posed}
	\end{align}
	where $u(x) = x + 1 + \epsilon \sin(\epsilon^{-2}x)$. 
Using several values of $\epsilon  > 0.66049142$, plot the corresponding density $a(x)$ for $x$ in `np.linspace(0,1,11)` to demonstrate that the problem is ill-posed.


```python
c0 = 1
for eps in [0.7, 0.8, 0.9, 1, 2]:
    x = np.linspace(0, 1, 11)
    du_func = lambda x: 1 + np.cos(x/eps**2)/eps
    du = du_func(x)
    If = -x
    def F(a):
        return np.sum(((c0 - If)/a - du)**2)
    guess = np.ones(len(x))*1/2
    a = minimize(F, guess).x
    plt.plot(x, a, label="eps="+str(eps))
    plt.title("eps="+str(eps))
    plt.show()
```
